```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
sales_monthly = pd.read_csv("vw_gold_sales_monthly_summary.csv")
sales_monthly['month'] = pd.to_datetime(sales_monthly['month'])
sales_monthly.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(sales_monthly['month'], sales_monthly['total_revenue'], marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(sales_monthly['month'], sales_monthly['total_orders'])
plt.title("Monthly Orders Count")
plt.xlabel("Month")
plt.ylabel("Orders")
plt.xticks(rotation=45)
plt.show()
```

```python
sales_region = pd.read_csv("vw_gold_sales_by_region.csv")
sales_region.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=sales_region, x='region', y='revenue')
plt.title("Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
headcount = pd.read_csv("vw_gold_hr_employee_headcount.csv")
headcount['month'] = pd.to_datetime(headcount['month'])
headcount.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(headcount['month'], headcount['headcount'], marker='o', color='purple')
plt.title("Employee Headcount Over Time")
plt.xlabel("Month")
plt.ylabel("Headcount")
plt.xticks(rotation=45)
plt.show()
```

```python
perf = pd.read_csv("vw_gold_product_performance.csv")
perf.head()
```

```python
top10 = perf.sort_values(by="total_revenue", ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top10, x="product_name", y="total_revenue")
plt.title("Top 10 Products by Revenue")
plt.xticks(rotation=75)
plt.ylabel("Revenue")
plt.show()
```

```python
top10_qty = perf.sort_values(by="total_quantity", ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top10_qty, x="product_name", y="total_quantity")
plt.title("Top 10 Products by Quantity Sold")
plt.xticks(rotation=75)
plt.ylabel("Quantity")
plt.show()
```

```python
clv = pd.read_csv("vw_gold_customer_lifetime.csv")
clv.head()
```

```python
plt.figure(figsize=(10,5))
plt.hist(clv['total_spend'], bins=20, color='green')
plt.title("Distribution of Customer Lifetime Spending")
plt.xlabel("Total Spend")
plt.ylabel("Number of Customers")
plt.show()
```
